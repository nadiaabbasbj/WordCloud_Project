```python
# Import necessary libraries
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
```

```python
#Mount Google Drive:
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Load the text file containing the data for the word cloud
with open("/content/drive/My Drive/WordCloud/sample_text.txt", "r") as file:
    text = file.read()
```

```python
# Create the WordCloud object with specified parameters
wordcloud = WordCloud(
width=800,
height=400,
background_color="white",
stopwords=set(STOPWORDS),
min_font_size=10,
).generate(text)
```

```python
# Display the generated word cloud using matplotlib
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```

```python
#Customization:
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color="black",
    stopwords=set(STOPWORDS).union({"example", "word"}),
    min_font_size=10,
    max_words=100,
    colormap='viridis'
).generate(text)
```

```python
# Display the generated customized word cloud using matplotlib
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```

```python
#Using a Mask
from PIL import Image
import numpy as np

mask_image = np.array(Image.open("/content/drive/My Drive/WordCloud/mask_image.jpeg"))
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color="white",
    mask=mask_image,
    stopwords=set(STOPWORDS),
    min_font_size=10,
).generate(text)
```

```python
# Display the word cloud
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud, interpolation='bilinear')  # Use bilinear interpolation for smoother display
plt.axis("off")  # Turn off the axis
plt.tight_layout(pad=0)  # Adjust layout
plt.show()  # Show the plot
```

```python
#loading the text file,
#generating the word cloud with the mask, and displaying it:

# Import necessary libraries
from wordcloud import WordCloud, STOPWORDS
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Load the text file
with open("/content/drive/My Drive/WordCloud/sample_text.txt", "r") as file:
    text = file.read()

# Load the mask image
mask_image = np.array(Image.open("/content/drive/My Drive/WordCloud/mask_image.jpeg"))

# Create the WordCloud object
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color="white",
    mask=mask_image,
    stopwords=set(STOPWORDS),
    min_font_size=10,
).generate(text)

# Display the word cloud
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
